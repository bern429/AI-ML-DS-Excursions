##### Load modules and packages.

```python
from google.colab import drive
import autosklearn.classification
import pandas as pd
```

##### Mount Google Drive to Colab.

```python
drive.mount('/content/gdrive')
```

##### Load the train and test sets from Google Drive.

```python
trainSet = pd.read_csv("gdrive/MyDrive/Colab Notebooks/4_heartDiseasePrediction/train.csv")
trainSet.head()
```

```python
testSet = pd.read_csv("gdrive/MyDrive/Colab Notebooks/4_heartDiseasePrediction/test.csv")
testSet.head()
```

##### Creat models, fit them to the training set, and choose the best model.

```python
automl = autosklearn.classification.AutoSklearnClassifier(time_left_for_this_task = 900, ensemble_size = 1)
xTrain = trainSet.drop(columns = ['target'])
yTrain = trainSet[['target']]
automl.fit(xTrain, yTrain)
```

##### Display all the models trained during the search process.

```python
automl.leaderboard(ensemble_only = False, sort_by = 'cost', sort_order = 'ascending')
```

```python
automl.show_models()
```

##### Predict on test set

```python
xTest = testSet.drop(columns = ['target'])
testSetPredictions = automl.predict(xTest)
testSetPredictions
```
