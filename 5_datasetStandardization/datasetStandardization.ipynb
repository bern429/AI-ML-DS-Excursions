##### **Standardize a dataset using basic Pandas operations or using the fit, transform, and fit_transform methods of sklearn.preprocessing.StandardScaler().**
##### $\rule{10.5in}{1pt}$

##### Load modules and packages.

```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
```

##### Create a dataset for demonstration.

```python
np.random.seed(1)
dataset = pd.DataFrame(np.round(100*np.random.rand(10, 3), 2), columns = ['x1', 'x2', 'x3'])
dataset
```

$\hspace{1in}$

##### **Standardize the dataset using basic Pandas operations.**

##### Calculate the mean of each column.

```python
columnMeans = dataset.mean(axis = 0)
columnMeans
```

##### Calculate the standard deviation of each column.

```python
columnSds = dataset.std(axis = 0, ddof = 0)
columnSds
```

##### Standardize each column.

```python
standardizedDataset_v1 = (dataset - columnMeans)/columnSds
standardizedDataset_v1
```

$\hspace{1in}$


##### **Standardize the dataset using the fit and transform methods of sklearn.preprocessing.StandardScaler().**

##### Instantiate the StandardScaler class and fit the instance to the dataset.

```python
scaler = StandardScaler()
scaler.fit(dataset)
```

##### Display the calculated mean of each column.

```python
scaler.mean_
```

##### Display the calculated standard deviation of each column.

```python
scaler.scale_
```

##### Apply the transform method to standardize each column.

```python
standardizedDataset_v2 = pd.DataFrame(scaler.transform(dataset), columns = ['x1', 'x2', 'x3'])
standardizedDataset_v2
```

$\hspace{1in}$

##### **Standardize the dataset using the fit_transform method of sklearn.preprocessing.StandardScaler().**

##### Instantiate the StandardScaler class and apply the fit_transform method to the dataset.

```python
scaler = StandardScaler()
standardizedDataset_v3 = pd.DataFrame(scaler.fit_transform(dataset), columns = ['x1', 'x2', 'x3'])
standardizedDataset_v3
```
